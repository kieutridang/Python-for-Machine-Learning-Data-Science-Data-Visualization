# Seaborn

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
# Tải dữ liệu iris
iris = sb.load_dataset("iris")
iris.head()

# Tải dữ liệu tips
tips = sb.load_dataset("tips")
tips.head()
```

```python
# Xem các dataset:
print(sb.get_dataset_names())
```

## Các loại biểu đồ

### swarmplot
- Sử dụng sb.swarmplot(): vẽ một scatterplot phân loại với các điểm không chồng chéo.
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.swarmplot.html
- Sử dụng plt.show() để hiển thị biểu đồ

```python
# Construct iris plot
# x: dữ liệu hiển thị trên trục hoành, y: dữ liệu hiển thị trên trục tung, data: bộ dữ liệu
plot= sb.swarmplot(x="species", y="petal_length", data=iris) 
# Show plot
# set title for plot
plot.set_title("Iris' types", fontsize=12, color='red')
plt.show()
```

### stripplot
- Sử dụng sb.stripplot(): vẽ một scatterplot phân loại với các điểm chồng lên nhau.
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.stripplot.html
- Sử dụng plt.show() để hiển thị biểu đồ

```python
# Construct iris plot
# x: dữ liệu hiển thị trên trục hoành, y: dữ liệu hiển thị trên trục tung, data: bộ dữ liệu
plot= sb.stripplot(x="species", y="petal_length", data=iris) 
# Show plot

plt.show()
```

### Histogram
- Sử dụng sb.distplot(): vẽ một phân bố quan sát đơn biến
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
# Reset default params
sb.set()
sb.distplot(iris['petal_length'],kde = False, color="Red") # kde = False thì representation của kernel estimation plot sẽ không vẽ, chỉ vẽ histogram
plt.show()
```

### Boxplot
- sb.boxplot(): vẽ plot để hiển thị các phân phối liên quan đến category
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sb.boxplot(x = "species", y = "petal_length", data = iris)
plt.show()
```

### violin plot
- sb.violinplot(): violin plot là sự kết hợp giữa box plots và density plots. 
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
sb.violinplot(x = "species", y = "petal_length", data = iris,palette='husl')
plt.show()
```

### boxenplot
- Sử dụng sb.boxenplot(): tương tự như box plot nhưng thể hiện rõ hơn thông tin phân bố dữ liệu
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.boxenplot.html

```python
sb.boxenplot(x = "species", y = "petal_length", data = iris,palette='husl')
plt.show()
```

### Barplot
- Sử dụng sb.barplot(): vẽ plot để hiển thị dữ liệu dưới dạng khối hình chữ nhật
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
sb.barplot(x = "species", y = "petal_length", data = iris)
plt.show()
```

### Lineplot
- Sử dụng sb.lineplot(): vẽ plot để hiển thị dữ liệu dưới dạng line
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.lineplot.html

## Regression plot


### regplot
- regplot tạo ra một scatter plot với một regression line

- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.regplot.html

```python
iris.info()
```

```python
iris.corr()
```

```python
sb.regplot(x='petal_length',y='sepal_length', data=iris) #chuyển sang polynomial regression sử dụng tham số “order”, vd: order=2

plt.show()
```

```python
# x_bins có thể được sử dụng để chia dữ liệu thành các phần rời rạc (discrete bins)
sb.regplot(x='petal_length',y='sepal_length', data=iris, x_bins=5) 

plt.show()
```

### residplot
- Đánh giá hồi quy với residplot, giá trị phần dư trên trục tung - giá trị x trên trục hoành.
- Nếu các trị phần dư phân tán ngẫu nhiên quanh giá trị zero (đường ngang) thì coi như phân tích hồi quy phù hợp
https://seaborn.pydata.org/generated/seaborn.residplot.html

```python
sb.residplot(x='petal_length',y='sepal_length', data=iris)
#có thể đánh giá polynomial regression bằng cách sử dụng tham số “order”, vd: order=2
```

### joinplot:

- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
sb.set()
# kind: 'scatter', 'reg', 'resid', 'kde', or 'hex'
sb.jointplot(x = 'petal_length',y = 'petal_width',data = iris, kind = 'reg', color='red') 
plt.show()
```

### lmplot
- Trong khi regplot () luôn hiển thị mối quan hệ x-y đơn giản, lmplot() kết hợp regplot() với FacetGrid cho phép hiển thị hồi quy tuyến tính trên các biểu đồ với nhiều tham số khác
- Xem thông tin chi tiết của function: https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
sb.lmplot(x='sepal_length',y='petal_length', col ='species', data=iris)
plt.show()
```

### Heatmap
- Heatmap là một biểu đồ ma trận phổ biến được sử dụng để tóm tắt mối quan hệ giữa hai biến.
- heatmap() yêu cầu dữ liệu phải ở định dạng lưới.

```python
# Create a crosstab table of the data
plt.figure(figsize=(12,4))
pd_crosstab = pd.crosstab(tips["time"], tips["size"],values=tips['total_bill'], aggfunc='sum')
#print(pd_crosstab)
pd_crosstab

# Plot a heatmap of the table
sb.heatmap(pd_crosstab)

plt.show()
```

```python
# Create a crosstab table of the data
plt.figure(figsize=(12,4))
pd_crosstab = pd.crosstab(tips["time"], tips["size"],values=tips['total_bill'], aggfunc='sum')
#print(pd_crosstab)
pd_crosstab

# Plot a heatmap of the table with color bar and reverse color
# The default cmap is sns.cm.rocket. To reverse it set cmap to sns.cm.rocket_r

sb.heatmap(pd_crosstab, cmap = sb.cm.rocket_r) 

plt.show()
```

```python
# Create a crosstab table of the data
plt.figure(figsize=(12,4))
pd_crosstab = pd.crosstab(tips["time"], tips["size"],values=tips['total_bill'], aggfunc='sum')
#print(pd_crosstab)
pd_crosstab

# Plot a heatmap of the table
sb.heatmap(pd_crosstab, cmap="YlGnBu") # có thể sử dụng palette màu tự định nghĩa cmap="YlGnBu"

plt.show()
```

```python

plt.figure(figsize=(10,4))

# Plot a heatmap of the correlation of dataframe
sb.heatmap(tips.corr()) 

plt.show()
```

## Vẽ biểu đồ trên Data Aware Grid


### FacetGrid

```python
g = sb.FacetGrid(tips, row='smoker', col='time')
g = g.map(plt.hist, 'total_bill')
```

```python
g = sb.FacetGrid(tips, row='smoker', col='time', row_order=['No','Yes'], col_order=['Dinner','Lunch'])
g = g.map(sb.regplot, 'total_bill', 'tip')
```

### factorplot - catplot

```python
sb.catplot(x="day", y="total_bill", hue="smoker", col="time", kind='swarm',data=tips)
plt.show()
```

### Pair grid
- Vẽ một mạng lưới các ô con bằng cách sử dụng cùng một loại biểu đồ để trực quan hóa dữ liệu.

```python
g = sb.PairGrid(iris)
g.map(plt.scatter)
plt.show()
```

```python
g = sb.PairGrid(iris)
g.map_diag(plt.hist) # trên đường chéo chính
g.map_offdiag(plt.scatter)
```
