# Chapter 5 - Exercise 1: GroupBy

#### drinks.csv là tập tin cung cấp dữ liệu về tình hình tiêu thụ rượu bia ở các quốc gia theo từng châu lục

country: tên quốc gia tiêu thụ<br/>
beer_servings: số lượng bia tiêu thụ<br/>
spirit_servings: số lượng rượu mạnh (>= 40<sup>0</sup>)<br/>
wine_servings: số lượng rượu vang<br/>
total_litres_of_pure_alcohol: tổng số lít rượu nguyên chất<br/>
continent: châu lục

```python
import pandas as pd
```

```python
# Câu 1: Đọc dữ liệu từ tập tin drinks.csv với index_col là cột đầu tiên của dữ liệu, và lưu vào biến drink. 
drinks = pd.read_csv('drinks.csv', index_col=0)
# Cho biết kiểu dữ liệu (type) của drink
print(type(drinks))
# Cho biết ra kích thước (shape) của drink 
print(drinks.shape)
# Hiển thị tên các cột (columns) của drink
print(drinks.columns)
```

```python
# Xem 5 dòng dữ liệu đầu tiên (head) của drink
drinks.head()
```

```python
# Xem 5 dòng dữ liệu cuối cùng (tail) của drink
drinks.tail()
```

```python
# Câu 2: Cho biết số lượng bia tiêu thụ trung bình ở mỗi châu lục
drinks.groupby('continent').beer_servings.mean()
```

```python
# Câu 3: Cho biết thông tin thống kê tổng quát (describe) số lượng rượu vang được tiêu thụ ở mỗi châu lục 
drinks.groupby('continent').wine_servings.describe()
```

```python
# Câu 4: Cho biết số lượng các loại bia và rượu tiêu thụ trung bình (mean) ở mỗi châu lục
drinks.groupby('continent').mean()
```

```python
# Câu 5: Cho biết giá trị trung vị (median) của các loại bia và rượu tiêu thụ ở mỗi châu lục
drinks.groupby('continent').median()
```

```python
# Câu 6: Cho biết số lượng rượu mạnh (spirit_servings) tiêu thụ trung bình, lớn nhất và nhỏ nhất ở mỗi châu lục 
# Gợi ý: Dùng spirit_servings.agg(['mean', 'min', 'max'])
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```

```python
# Câu 7: Sắp xếp dữ liệu tăng dần (sort_values) theo số lượng bia tiêu thụ
drinks_des = drinks.sort_values(by='beer_servings', ascending=True)
drinks_des
```

```python
# Cho biết 5 quốc gia có lượng tiêu thụ bia lớn nhất
drinks_des.tail()
```

```python
# Cho biết 5 quốc gia có lượng tiêu thụ bia nhỏ nhất
drinks_des.head()
```
