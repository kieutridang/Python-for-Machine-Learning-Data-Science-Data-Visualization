```python
# Nếu có, hãy thay thế với quy tắc sau:
# Nếu Null cột 'high' thì thay bằng giá trị max trên cột 'high' của mã chứng khoán đó
# Nếu Null cột 'low' thì thay bằng giá trị min trên cột 'low' của mã chứng khoán đó
# HINT: stocks1['high'].replace(np.nan, stocks1.groupby(['symbol'])['high'].transform(max), inplace=True)
replaced_high_value = stocks1.groupby(['symbol'])['high'].transform(max)
replaced_low_value = stocks1.groupby(['symbol'])['low'].transform(min)

stocks1['high'].replace(np.nan, replaced_high_value, inplace = True)
stocks1['low'].replace(np.nan, replaced_low_value, inplace = True)

stocks1
```
