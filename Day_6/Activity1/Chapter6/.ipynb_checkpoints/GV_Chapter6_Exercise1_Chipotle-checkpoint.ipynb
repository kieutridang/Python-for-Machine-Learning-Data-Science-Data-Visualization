## Chapter 6 - Exercise 1: Trực quan hóa dữ liệu Chipotle

### Cho dữ liệu 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

Hướng dẫn chung:
1. Gán dữ liệu vào biến:  <p style="color:blue"> biến = pd.read_csv(url, sep = '\t')</p>
2. Để tạo một dictionary với key là item_name và value là tần suất xuất hiện của item_name: <span style="color:blue">collections.Counter(biến_chứa_các_item_name)</span>
3. Chuyển dictionary thành dataframe: <span style="color:blue">pd.DataFrame.from_dict(dictionary, orient='index')</span>

#### Thực hiện các yêu cầu sau, và đối chiếu với kết quả cho trước:

```python
import pandas as pd
import collections
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
# Câu 1: Đọc dữ liệu và gán vào cho biến chipo. Hiển thị 10 dòng đầu của dữ liệu
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'  
chipo = pd.read_csv(url, sep = '\t')
chipo.head(10)
```

```python
# Câu 2: 
# Tạo biến x chứa các item_name, in head của x
x = chipo.item_name
print(x.head())

# Sử dụng collections.Counter(x) để tạo một dictionary với key là item_name và value là tần suất xuất hiện của item_name, in kết quả
letter_counts = collections.Counter(x)
print(letter_counts)
```

```python
# Câu 3: 
# Chuyển dictionary câu 2 thành DataFrame df
df = pd.DataFrame.from_dict(letter_counts, orient='index')
print(df.head())
```

```python
# Câu 4:
# a) Sắp xếp df theo tần suất giảm dần, và lấy 5 item đầu tiên
df_5 = df.sort_values(by = 0, ascending = False)[0:5]
df_5
```

```python
# b) Vẽ biểu đồ khối với 5 item này có title xlabel, ylabel và xsticks

# Tạo plot
plt.bar(df_5.index.values, df_5[0].values)

# Thiết lập title và labels, xsticks (rotation='vertical')
plt.xlabel('Items')
plt.ylabel('Price')
plt.title('Most ordered Chipotle\'s Items')
plt.xticks(df_5.index.values, df_5.index.values, rotation='vertical')

# Hiển thị biểu đồ
plt.show()
```
