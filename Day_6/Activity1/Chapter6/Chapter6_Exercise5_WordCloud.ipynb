## Chapter 6 - Exercise 5: WordClouds

### Part 1:

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# Tạo dataframe df từ file dữ liệu Youtube04-Eminem.csv, in vài dòng dữ liệu đầu của df
df = pd.read_csv("data\Youtube04-Eminem.csv", encoding ="latin-1") 
df.head()
df['CONTENT'].head(20)
```

```python
# Hãy bỏ các STOPWORD
stopwords = set(STOPWORDS) 
stopwords
```

```python
comment_words =' '.join(df['CONTENT'])
```

```python
# Vẽ WordClouds cho dữ liệu này
wc = WordCloud(
    background_color='black',
    stopwords=stopwords
)

# generate the word cloud
wc.generate(comment_words)

# display the word clouds
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Part 2: 

```python
# Sử dụng vietnamese stop word từ : https://github.com/stopwords/vietnamese-stopwords => tạo thành set các stop words
stopwords = set()
f = open(r"data\vietnamese-stopwords.txt", "r", encoding='utf-8')
for line in f:
    word = f.readline()
    stopwords.add(word.replace('\n',''))
f.close()

stopwords
```

```python
# Đọc file ngon_tu_quang_cao.txt => đưa nội dung vào biến text

text = ''
with open(r'data\ngon_tu_quang_cao.txt', 'r', encoding='utf-8') as f:
    text = f.read()
text
```

```python
# Bổ sung thêm một số từ không quan trọng vào stopwords
list_of_words = ['và', 'một', 'của', 'có', 'đó', 'rất', 'nào', 'được',
                'khi', 'thể', 'sự', 'tính', 'trong','cũng','cùng','cho','hay','chỉ']
for word in list_of_words:
    stopwords.add(word) 
```

```python
# Vẽ wordclouds
wc = WordCloud(
    background_color='black',
    max_words=1000,
    stopwords=stopwords
)

wc.generate(text)

plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
# Chọn hình làm wc_mask phù hợp
wc_mask = np.array(Image.open('ad_s.png'))

plt.imshow(wc_mask, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Vẽ wordclouds với wc_mask
wc1 = WordCloud(background_color='white', max_words=1000, 
                mask=wc_mask, stopwords=stopwords)

wc1.generate(text)

plt.figure(figsize=(10, 12))
plt.imshow(wc1, interpolation='bilinear')
plt.axis('off')
plt.show()
```
