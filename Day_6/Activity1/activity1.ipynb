## Visualizing stock trends.

Giả sử bạn muốn đầu tư vào chứng khoán và bạn đã download giá đóng cửa - mở cửa của 5 "ông lớn" sau: Amazon, Google, Apple, Facebook, and Microsoft. 
Hãy tạo đồ thị line để quan sát xu hướng chứng khoán của các công ty trên. 


```python
# Import thư viện
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Sử dụng thư viện pandas để đọc dữ liệu

```python
# load dữ liệu
amazon = pd.read_csv('./data/AMZN_data.csv')
google = pd.read_csv('./data/GOOGL_data.csv')
apple = pd.read_csv('./data/AAPL_data.csv')
facebook = pd.read_csv('./data/FB_data.csv')
microsoft = pd.read_csv('./data/MSFT_data.csv')
amazon.sort_values('close', ascending=False).head()
```

Sử dụng Matplotlib để tạo một biểu đồ trực quan hóa giá đóng cửa trong 5 năm qua (toàn bộ dữ liệu trong file) cho cả 5 công ty. Lưu ý có nhãn, tiêu đề và legend để làm cho đồ thị dễ xem hơn và sử dụng plt.grid () cho đồ thị.

```python
# Tạo figure (figure size of 16"x8")
plt.figure(figsize=(16, 8))
# Vẽ đồ thị
plt.plot(amazon['date'], 'close', data=google, label='Google')
plt.plot(amazon['date'], 'close', data=facebook, label='Facebook')
plt.plot(amazon['date'], 'close', data=microsoft, label='Microsoft')
plt.plot(amazon['date'], 'close', data=apple, label='Apple')
plt.plot(amazon['date'], 'close', data=amazon, label='Amazon')
# Đặt ticks cho trục x- và trục y
plt.xticks(np.arange(0, 1260, 40), rotation=70)
plt.yticks(np.arange(0, 1450, 100))
# Title and labels
plt.title("Giá đóng cửa trong 5 năm")
#  grid
plt.grid()
#  legend
plt.legend(['Google', 'Facebook', 'Microsoft', 'Apple', 'Amazon'])
# Show plot
plt.show()
```
