```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python

fig, ax = plt.subplots()
ax.plot(x, y)
# Gán nhãn cho biểu đồ 

plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
f.set_size_inches(14,4)

ax1.plot(x, y, color='r')
ax2.scatter(x, y, color='g')

# Hiển thị tiêu đề cho biểu đồ
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True)
fig.set_size_inches(12,6)
axes[0, 1].plot(x, y, color='r')
# vẽ ở các biểu đồ khác màu khác nhau/style marker khác nhau
plt.show()
```

```python
x= [1.5, 2.5, 3.5]
y= [2, 3, 4]
fig2 = plt.figure()
fig2.set_size_inches(12,6)
spec2 = gridspec.GridSpec(ncols=2, nrows=2)
f2_ax1 = fig2.add_subplot(spec2[0, 0])
f2_ax2 = fig2.add_subplot(spec2[0, 1])
f2_ax3 = fig2.add_subplot(spec2[1, 0])
f2_ax4 = fig2.add_subplot(spec2[1, 1])
fig2.tight_layout()

f2_ax1.plot([1, 2, 3], label='line1', marker='*', markersize=10)
# Vẽ ở các biểu đồ con các loại biểu đồ khác nhau (bar, plot,...)

plt.show()
```
