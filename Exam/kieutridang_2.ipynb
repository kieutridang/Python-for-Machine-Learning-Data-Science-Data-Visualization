```python
# b) Hiển thị tỷ lệ thất nghiệp trung bình các tháng trong năm 2016 rồi nhận xét, kết quả gợi ý như hình sau
import datetime as dt
from datetime import datetime

Months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(12,6))
sns.pointplot(data=us_unemployment, y='Rate', x='Month', sort=True, order=Months)
```
